```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.tsa.ar_model import AutoReg
import math
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15,6
import warnings
warnings.filterwarnings('ignore')
```

```python
data= pd.read_csv('Demand_History.csv')
```

```python
data_copy = data.copy()
data.drop(columns=['Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14'], axis = 1, inplace = True)
data_copy.rename(columns = {'Unnamed: 12': '2019', 'Unnamed: 13': '2020'}, inplace = True)
data_copy.drop(columns = 'Unnamed: 14', axis = 1, inplace = True)
```

```python
data_copy
```

```python
for row in range(len(data)):
  data_0 = data[row:row+1]
  data_0.drop(columns=['demand_point_index', 'x_coordinate', 'y_coordinate'], axis = 1, inplace = True)
  data_0 = data_0.transpose()
  data_0['Demand'] = data_0[row]
  data_0.drop(columns = row, inplace=True)
  ts = data_0['Demand']
  ts_log = np.log(ts)
  try:
    model = AutoReg(ts_log, lags = 1).fit()
    predict = model.predict(start = len(ts_log), end = len(ts_log) + 1 )
    pred_original = np.exp(predict)
    data_copy['2019'][row] = pred_original[0]
    data_copy['2020'][row] = pred_original[1]
  except:
    data_copy['2019'][row] = 0
    data_copy['2020'][row] = 0
```
